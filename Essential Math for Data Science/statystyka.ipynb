```python
probka = [0, 1, 5, 7, 9.5, 10, 14]

def mediana(probka):
  uporzadkowanie = sorted(probka)
  n = len(probka)
  srodek = int(n / 2) - 1 if n % 2 == 0 else int(n / 2)
  print(f'Środek: {srodek+1}')

  if n % 2 == 0:
    print('Liczba próbek jest parzysta.')
    print((uporzadkowanie[srodek] + uporzadkowanie[srodek+1]) / 2.0)
  else:
    print('Liczba próbek jest nieparzysta.')
    print(uporzadkowanie[srodek])

mediana(probka)
```

```python
from collections import defaultdict

probka = [1, 3, 2, 5, 7, 0, 2, 3]

def dominanta(wartosci):
  wystapienia = defaultdict(lambda: 0)

  for s in wartosci:
    wystapienia[s] += 1

  najwiecej_wystapien = max(wystapienia.values())
  dominanty = [v for v in set(wartosci) if wystapienia[v] == najwiecej_wystapien]
  return dominanty

print(dominanta(probka))
```

```python
from math import sqrt

dane = [0, 1, 5, 7, 9, 10, 14]

def wariancja(wartosci):
  srednia = sum(wartosci) / len(wartosci)
  wariancja_ = sum([(i - srednia)**2 for i in wartosci]) / len(wartosci)

  return sqrt(wariancja_)

wariancja(dane)
```

```python
from math import sqrt

dane = [0, 1, 5, 7, 9, 10, 14]

def wariancja(wartosci, czy_probka=False):
  srednia = sum(wartosci) / len(wartosci)
  wariancja_ = sum([(i - srednia)**2 for i in wartosci]) / (len(wartosci) - (1 if czy_probka else 0))

  return sqrt(wariancja_)

wariancja(dane, czy_probka=True)
```

```python
import math

def rozklad_normalny(x: float, srednia: float, odch_std: float) -> float:
  return (1.0 / (2.0 * math.pi * odch_std ** 2) ** 0.5) * math.exp(-1.0 * ((x-srednia)** 2 / (2.0 * odch_std ** 2)))
```

```python
from math import sqrt
from scipy.stats import norm

def krytyczna_wartosc_z(p):
  norm_dist = norm(loc=0.0, scale=1.0)
  lewy_ogon = (1.0 - p) / 2.0
  prawy_ogon = 1.0 - ((1.0 - p) / 2)
  return norm_dist.ppf(lewy_ogon), norm_dist.ppf(prawy_ogon)

def przedzial_ufnosci(p, srednia_proby, odch_proby, n):

  dolna, gorna = krytyczna_wartosc_z(p)
  dolna_ci = dolna * (odch_proby / sqrt(n))
  gorna_ci = gorna * (odch_proby / sqrt(n))

  return srednia_proby + dolna_ci, srednia_proby + gorna_ci

przedzial_ufnosci(p=0.95, srednia_proby=29.214, odch_proby=1.37, n=31)
```

```python
#Ćwiczenie 1
import math

def odch_std_oraz_srednia(elementy):
  srednia = sum(elementy) / len(elementy)
  sigma = 0

  for element in elementy:
    sigma += (element - srednia)**2

  return (srednia, math.sqrt(sigma / len(elementy)))

odch_std_oraz_srednia([1.78, 1.75, 1.72, 1.74, 1.77])
```
