```python
#Łączenie dwóch przekształceń
import numpy as np

i_daszek1 = np.array([0, 1])
j_daszek1 = np.array([-1, 0])
przeksztalcenie1 = np.array([i_daszek1, j_daszek1]).transpose()

i_daszek2 = np.array([1, 0])
j_daszek2 = np.array([1, 1])
przeksztalcenie2 = np.array([i_daszek2, j_daszek2]).transpose()

polaczone = przeksztalcenie2 @ przeksztalcenie1

print('POŁĄCZONA MACIERZ:\n {}'.format(polaczone))

v = np.array([1, 2])
print(polaczone.dot(v))
```

```python
#Rozkład własny macierzy
from numpy import array, diag
from numpy.linalg import eig, inv

A = array([
    [1, 2],
    [4, 5]
])

wart_wlasne, wekt_wlasne = eig(A) #Rozkład macierzy A na czynniki

print('WARTOSCI WLASNE')
print(wart_wlasne)
print('WEKTORY WLASNE')
print(wekt_wlasne)
```

```python
#Rozkład własny macierzy
from numpy import array, diag
from numpy.linalg import eig, inv

A = array([
    [1, 2],
    [4, 5]
])

wart_wlasne, wekt_wlasne = eig(A) #Rozkład macierzy A na czynniki

print('WARTOSCI WLASNE')
print(wart_wlasne)
print('WEKTORY WLASNE')
print(wekt_wlasne)

print('\nODTWORZONA MACIERZ')
Q = wekt_wlasne
R = inv(Q)

L = diag(wart_wlasne)
B = Q @ L @ R #wymnażanie(Rekonstrukcja do pierwotnej macierzy A)

print(B)
```

#Ćwiczenia

```python
#Ćwiczenie 1
import numpy as np

i_daszek = np.array([2, 0])
j_daszek = np.array([0, 1.5])

baza = np.array([i_daszek, j_daszek]).transpose()

vector = np.array([1, 2])

nowy_vector = baza.dot(vector)
nowy_vector
```

```python
#Ćwiczenie 2
import numpy as np

i_daszek = np.array([-2, 1])
j_daszek = np.array([1, -2])

baza = np.array([i_daszek, j_daszek]).transpose()

vector = np.array([1, 2])

nowy_vector = baza.dot(vector)

print(nowy_vector)
```

```python
#Ćwiczenie 3
import numpy as np
from numpy.linalg import det

i_daszek = np.array([1, 0])
j_daszek = np.array([2, 2])

baza = np.array([i_daszek, j_daszek]).transpose()

wyznacznik = det(baza)

#Wyznacznik jest to wartość o którą pomnożył się obszar zajęty przez wektory
#Pierwotnie wektory i oraz j zajmowały obszar o wyznaczniku 1(potocznie mówiąc jedną kratkę)
#Gdy wartość wektora j zmieniła się na [2, 2] to wartość wyznacznika się podwoiła
#Obecnie wynosi 4 połowy kratki(gdyż doszło do ścięcia) co ostatecznie daje wartość 4*1/2 = 2
wyznacznik
```

```python
#Ćwiczenie 5
import numpy as np
from numpy.linalg import inv

A = np.array([
    [3, 1, 0],
    [2, 4, 1],
    [3, 1, 8]
])

B = np.array([
    54,
    12,
    6
])

rozwiazanie = inv(A).dot(B)

rozwiazanie
```
